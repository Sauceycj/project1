```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("AviationData.csv", encoding="ISO-8859-1")
df.head()
```

```python
# Data cleaning
```

```python
df = df.dropna(subset=[
    "Aircraft.Category", "Make", "Purpose.of.flight", "Number.of.Engines" 
])
print("Remaining rows:", df.shape[0])               
```

```python
text_columns = ["Aircraft.Category", "Make", "Purpose.of.flight", "Number.of.Engines"]



for col in text_columns:
    df[col] = df[col].astype(str)         
    df[col] = df[col].str.strip()         
    df[col] = df[col].str.title()   

df[["Aircraft.Category", "Make", "Purpose.of.flight", "Number.of.Engines"]].head(30)
```

```python
df["Event.Date"] = pd.to_datetime(df["Event.Date"], errors='coerce')  
df["Year"] = df["Event.Date"].dt.year     

df[["Event.Date", "Year"]].head()
```

```python
df[text_columns] = df[text_columns].fillna("Unknown")
df[["Aircraft.Category", "Make", "Purpose.of.flight", "Number.of.Engines"]].head(30)
```

```python
df["Total.Fatal.Injuries"] = df["Total.Fatal.Injuries"].fillna(0)
df["Total.Serious.Injuries"] = df["Total.Serious.Injuries"].fillna(0)
df["Total.Minor.Injuries"] = df["Total.Minor.Injuries"].fillna(0)

df["Total.Injuries"] = (
    df["Total.Fatal.Injuries"].fillna(0) +
    df["Total.Serious.Injuries"].fillna(0) +
    df["Total.Minor.Injuries"].fillna(0)
)

df[[
    "Total.Fatal.Injuries", 
    "Total.Serious.Injuries", 
    "Total.Minor.Injuries", 
    "Total.Injuries"
]].head(30)
```

```python
df = df.drop_duplicates()

print("Final cleaned dataset rows:", df.shape[0])
```

Exporting data for tableau

```python
df.to_csv("data_cleaned.csv", index=False)
```

# Data visulization

UNIVARIATE

```python
plt.figure(figsize=(8,5))
df["Aircraft.Category"].value_counts().plot(kind="bar", color="seagreen")

plt.title("Aircraft Category Distribution", fontsize=16)
plt.xlabel("Aircraft Category")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["Purpose.of.flight"].value_counts().plot(kind="bar", color="coral")

plt.title("Flight Purpose Distribution", fontsize=16)
plt.xlabel("Purpose of Flight")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

Personal flights are responsible for big portion of aviation accidents.

```python
accidents_per_year = df["Year"].value_counts().sort_index()
```

```python
plt.figure(figsize=(12,6))
accidents_per_year.plot(kind="line", color="steelblue", marker='o')

plt.title("Accidents Over the Years", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Accidents", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

BIVARIATE

```python
top_makes = df["Make"].value_counts().head(10)

plt.figure(figsize=(10,6))
top_makes.plot(kind="bar", color="royalblue")
plt.title(" MAKE OF PLANE BY NUMBER OF ACCIDENTS", fontsize=16)
plt.ylabel("Number of Incidents")
plt.xlabel("Aircraft Make")
plt.tight_layout()
plt.show()
```

```python

injuries_by_purpose = df.groupby("Purpose.of.flight")["Total.Injuries"].sum().sort_values(ascending=False)


plt.figure(figsize=(12,6))
injuries_by_purpose.plot(kind="bar", color="skyblue")
plt.title("TOTAL INJURIES BY PURPOSE OF FLIGHT", fontsize=20)
plt.xlabel("Purpose of Flight", fontsize=15)
plt.ylabel("Total Injuries", fontsize=15)
plt.tight_layout()
plt.show()
```

```python

injuries_by_category = df.groupby("Aircraft.Category")["Total.Injuries"].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
injuries_by_category.plot(kind="bar", color="coral")
plt.title("TOTAL INJURIES BY AIRCRAFT CATEGORY", fontsize=16)
plt.xlabel("Aircraft Category", fontsize=12)
plt.ylabel("Total Injuries", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
injuries_by_engines = df.groupby("Number.of.Engines")["Total.Injuries"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
injuries_by_engines.plot(kind="bar", color="skyblue")

plt.title("TOTAL INJURIES BY NUMBER OF ENJINES", fontsize=16)
plt.xlabel("Number of Engines", fontsize=12)
plt.ylabel("Total Injuries", fontsize=12)
plt.tight_layout()
plt.show()
```

LINE GRAPH

```python
top_makes_injuries = df.groupby("Make")["Total.Injuries"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,5))
top_makes_injuries.plot(kind="line", marker='s', color='darkorange')
plt.title("Top 10 Aircraft Makes by Total Injuries")
plt.xlabel("Aircraft Make")
plt.ylabel("Total Injuries")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
injuries_by_purpose = df.groupby("Purpose.of.flight")["Total.Injuries"].sum().sort_values()

plt.figure(figsize=(12,5))
injuries_by_purpose.plot(kind="line", marker='^', color='green')
plt.title("Total Injuries by Purpose of Flight")
plt.xlabel("Purpose of Flight")
plt.ylabel("Total Injuries")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
injuries_by_category = df.groupby("Aircraft.Category")["Total.Injuries"].sum().sort_values()

plt.figure(figsize=(10,5))
injuries_by_category.plot(kind="line", marker='o', color='royalblue')
plt.title("Total Injuries by Aircraft Category")
plt.xlabel("Aircraft Category")
plt.ylabel("Total Injuries")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
injuries_by_engines.plot(kind="line", marker="o", color="purple")

plt.title("Line Plot: Injuries by Number of Engines")
plt.xlabel("Number of Engines")
plt.ylabel("Total Injuries")
plt.grid(True)
plt.tight_layout()
plt.show()
```
